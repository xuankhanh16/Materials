```python
player_list = ["Alice", "Bob", "Charles", "Derek"]

## Your Code for Task 1.1 here ##

def play_game(player_list, n):
    #consider case where no player
    if len(player_list) ==0:
        return None
    #if n is negative -> reverse the list of players except for the initial person aka holding-bomb person
    if n<0:
        player_list = [player_list[0]] + player_list[len(player_list):0:-1]
        n *= -1
    current_pos = 0
    while(len(player_list)>1):
        eliminated = (current_pos + n) % len(player_list)
        print(f"{player_list[eliminated]} gets eliminated")
        player_list.pop(eliminated)
        current_pos = (eliminated) % len(player_list) #the same index eliminated will refer to the next person since an element pops out
        #current_pos = (current_pos+n+1) % len(player_list)
    print(f"{player_list[0]}")
    
play_game(player_list, 4)
```

```python
#Task 6

def same_type(L):
    if len(L) < 2: #[1]
        return True
    else:
        if type(L[0]) == type(L[1]): #[1]
            return same_type(L[1:]) #[2]
        else:
            return False #[1]

print(same_type([1, 2, 3]))           # True
print(same_type([1, "2", 3]))         # False
print(same_type(["a", "b", "c"]))     # True
print(same_type([]))
```
