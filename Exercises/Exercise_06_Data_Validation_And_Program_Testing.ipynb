## Exercise 6.5 ISBN 

The International Standard Book Number (ISBN) is a numeric commercial book identifier which is intended to be unique. Publishers purchase ISBNs from an affiliate of the International ISBN Agency.

Write a code that 
- takes in a string input,
- return `True` if the string is either a valid ISBN-10 or ISBN-13 numbers.

Each line in the text file, `ISBN_EXERCISE.TXT` under the folder `resources`, contain ISBN-10 or ISBN-13 which can be valid or invalid.

Write a program to:
- print out the valid ISBNs in the file.
- print out the number of valid ISBNs in the file.

````python
def checkISBN(user_input):
    try:
        sumdigit = 0
        user_input_clean = user_input.replace("-", "")

        if len(user_input_clean) == 13:  # ISBN-13
            for j in range(12):
                digit = int(user_input_clean[j])
                if j % 2 == 0:
                    sumdigit += digit
                else:
                    sumdigit += digit * 3
            checkdigit = (10 - sumdigit % 10) % 10
            return checkdigit == int(user_input_clean[12])

        elif len(user_input_clean) == 10:  # ISBN-10
            for i in range(9):
                sumdigit += int(user_input_clean[i]) * (i + 1)
            last_char = user_input_clean[9]
            if last_char == 'X':
                check_digit = 10
            else:
                check_digit = int(last_char)
            return sumdigit % 11 == check_digit

        else:
            return False

    except:
        return False


with open("resources/ISBN_EXERCISE.txt", "r") as f:
    cnt = 0
    valid_isbn = []
    data = f.readlines()
    data = [x.strip() for x in data]
    isbn_no = [line.split(": ")[1] for line in data]
    for line in isbn_no:
        if checkISBN(line):
            cnt += 1
            valid_isbn.append(line)
    
    print(cnt)
    print(valid_isbn)
````

A pixel holds the information of a color and the colors are often represented in a string starting with a hex `#` character followed by 3 pairs of hexadecimal digits. We will call this the hexadecimal representation of the pixel. For example, the color white is `#FFFFFF` and the color silver is `#C0C0C0`.

### Task 1
Write a program code with the following specification:
- input a hexadecimal number as a string
- validate the input
- calculate the denary value of the hexadecimal number input
- output the denary value.

````python
def validate_hex(hexstring):
    for s in hexstring:
        if s not in "0123456789abcdefABCDEF":
            return False
    return True

def hextodenary(hexstring):
    return int(hexstring, 16)

def convert(hexstring):
    RGB = []
    if len(hexstring) != 6:
        print("Hex string must be exactly 6 characters for RGB.")
        return None
    
    for i in range(0, 6, 2):  # step by 2
        val = hexstring[i:i+2]
        RGB.append(int(val, 16))  # convert each pair to decimal
    return RGB

# Main code
hexstring = input("Hexstring is: ").strip()

if validate_hex(hexstring):
    print("Valid hex:", hexstring)
    print("Decimal value of full hex:", hextodenary(hexstring))
    print("RGB values:", convert(hexstring))
else:
    print("Invalid hex string.")
````

Each pair of the hexadecimal digits actually represents the intensity of 3 primary colors, red, green and blue, that make up the color in the pixel. As such, instead of having the hexadecimal representation of the pixel, each pixel can be represented by a list of 3 denary values that represents the intensity of the colours, red, green and blue respectively, e.g., the color `#FFCO80` can be represented as the list `[255,192,128]`. We will call this representation of the pixel the `RGB representation`.

### Task 2
A 16 by 13 pixels image is stored as hexadecimal form in the text file `TASK3_2.CSV`, where each line represents a row in the image.
Write a program code with the following specification: 
- read the hexadecimal values from the file TASK3_2.CSV and store them in a suitable format
- using the program code in Task 3.1,
- convert the image from its hexadecimal pixels representation ito the RGB presentation,
- write the new presentation into the file `MY_RGB_IMAGE.TXT`. 
Conversely, sometimes pixels are given in their RGB representation instead and for graphic designers, hexadecimal values are often preferred.

````python
import csv

def hextodenary(hexstring):
    return int(hexstring, 16)

def convert(hexstring):
    hexstring = hexstring[1:]  # Removing the '#' if present (assuming hex colors like "#FFA07A")
    RGB = []
    for i in range(0, 6, 2):  # Extract two characters at a time
        val = hexstring[i:i+2]  # Get the two-character hex value
        valden = hextodenary(val)  # Convert hex to denary
        RGB.append(valden)
    return RGB  # Return the RGB values

# Read the CSV file with hex color values
with open("resources/TASK3_2.csv", "r") as f:
    x = csv.reader(f)
    data = [row[0] for row in x]  # Assuming each row contains one hex string in the first column

# Write the RGB values to a text file
with open("resources/MY_RGB_IMAGE.TXT", "a") as file:
    for hexstring in data:
        rgb_values = convert(hexstring)  # Convert hex to RGB
        file.write(f"{rgb_values}\n")  # Write the result to the file

print("Conversion complete!")
````

### Task 3
A 16 by 13 pixels image is stored its RGB representation in the text file `TASK3_3.CSV` in the `resources` folder, where each line represents a row in the image.
Write an additional program code with the following specification: 
- read the rgb values from the file `TASK3_3.CSV` and store them in a suitable format,
- write the new presentation into the file `MY_HEX_IMAGE.TXT`.

````python
import csv

def rgb_to_hex(rgb):
  """Converts a list of RGB values [R, G, B] to a hexadecimal color string #RRGGBB."""
  return '#{:02x}{:02x}{:02x}'.format(rgb[0], rgb[1], rgb[2])

# Path to the input and output files
input_file = 'resources/TASK3_3.CSV' # Assuming resources folder exists
output_file = 'MY_HEX_IMAGE.TXT'

hex_image_data = []

try:
    with open(input_file, 'r') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            hex_row = []
            # Assuming each row contains RGB values for multiple pixels
            # and each pixel has 3 RGB values.
            # You might need to adjust the step based on the actual file format.
            for i in range(0, len(row), 3):
                try:
                    # Convert RGB values to integers
                    rgb_values = [int(row[i]), int(row[i+1]), int(row[i+2])]
                    hex_color = rgb_to_hex(rgb_values)
                    hex_row.append(hex_color)
                except (ValueError, IndexError) as e:
                    print(f"Error processing RGB values in row: {row}. Error: {e}")
                    # Handle potential errors in data format
                    continue
            hex_image_data.append(hex_row)

except FileNotFoundError:
    print(f"Error: The file {input_file} was not found.")
except Exception as e:
    print(f"An error occurred while reading the file: {e}")

# Write the hexadecimal data to the output file
try:
    with open(output_file, 'w') as outfile:
        for row in hex_image_data:
            outfile.write(','.join(row) + '\n') # Use comma as a delimiter

except Exception as e:
    print(f"An error occurred while writing to the file: {e}")
    
````
