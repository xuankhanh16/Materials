## Exercise 6.4

The file `WORDS.TXT` under the folder `resources` contains a list of single word computing terms used in a textbook.
	
Each entry has the following format:
>```python
><computing term>
><number>
>```

One of the file entries (in both files) is:
>```python
>program	
>52
>```
	
This means that after a complete scan of the textbook the word `program` was found 52 times.
	
By utilising the `try-except` statement, write program code to find and output the term with the highest number of occurrences.



````python
# YOUR ANSWER HERE

with open ("resources/WORDS.TXT", "r") as f:
    x = f.readlines()
    x_clean = [lines.strip() for lines in x]
    #print(x_clean)
    #print(x_clean[1])
    
term_count = {}

for i in range (0, len(x_clean), 2):
    try:
        terms = x_clean[i]
        count = int(x_clean[i+1])
        term_count[terms]=count
    except ValueError:
        print(f"invalid entry at line {i} and {i+1}")

most_common_term = max(term_count, key=lambda item:item[1])
print(most_common_term)
print(term_count[most_common_term])
````

## Exercise 6.5 ISBN 

The International Standard Book Number (ISBN) is a numeric commercial book identifier which is intended to be unique. Publishers purchase ISBNs from an affiliate of the International ISBN Agency.

Write a code that 
- takes in a string input,
- return `True` if the string is either a valid ISBN-10 or ISBN-13 numbers.

Each line in the text file, `ISBN_EXERCISE.TXT` under the folder `resources`, contain ISBN-10 or ISBN-13 which can be valid or invalid.

Write a program to:
- print out the valid ISBNs in the file.
- print out the number of valid ISBNs in the file.

````python
def checkISBN(user_input):
    try:
        sumdigit = 0
        user_input_clean = user_input.replace("-", "")
        if (len(user_input_clean)==13):
            for j in range (12):
                digit = int(user_input_clean[j])
                if (j%2==0):
                    sumdigit=sumdigit + digit
                else:
                    sumdigit = sumdigit + digit*3
            checkdigit = (10-sumdigit%10)%10
                
            return checkdigit == int(user_input_clean[12])
               

        elif (len(user_input_clean)==10):
            for i in range (9):
                sumdigit += int(user_input_clean[i]) * (i+1)
            return sumdigit%11 == int(user_input_clean[9])
        else:
            return False
    
    except:
        return False

    
with open("resources/ISBN_EXERCISE.txt", "r") as f:
    cnt = 0
    valid_isbn = []
    data = f.readlines()
    data = [x.strip() for x in data]
    isbn_no = [line.split(": ")[1] for line in data]
    for line in isbn_no:
        if checkISBN(line):
            cnt += 1
            valid_isbn.append(line)
    
    print(cnt)
    print(valid_isbn)


        
````
