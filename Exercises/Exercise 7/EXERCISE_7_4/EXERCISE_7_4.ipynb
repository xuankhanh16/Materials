```python
def getinfo():
    screen_name = []
    people = []
    with open('people.txt', "r") as file:
        for line in file:
            name, dob, identity = line.strip().split(',')
            cleaned_name = name.replace(" ", "")
            dob = dob[5:7]+dob[8:10]
            screen_name = cleaned_name + dob
            people.append({
                'fullname': name,
                'screenname': screen_name,
                'identity': identity
            })
    return people


# for person in people:
#     print(person['screenname'])


from flask import Flask, render_template
app = Flask(__name__)

@app.route('/')
def display_people():
    people = getinfo()
    return render_template ('display.html', people = people)


app.run()
    

        
```
