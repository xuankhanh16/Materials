# 2024/JPJC Promo P2/Q1
JP Bookstore maintains a list of books in a text file called TASK1.txt. Each line
in the file contains information about a single book in the following format:

```text
<Title>,<Genre>,<Price>
```

## Task 1.1

Write a function `read_books_data()` to:
- read the contents of TASK1.txt
- return a list of lists.

Each inner list should represent a book with elements: title, genre, and price with
their appropriate data types. <div style="text-align: right">[4]</div>


Test your program and show the returned list of lists. <div style="text-align: right">[1]</div>

````python
# Task 1.1


def read_books_data():
    Book = []
    with open ("TASK1.txt") as f:
        for line in f:
            elements = line.strip().split(',')
            Book.append(elements)
    return Book


read_books_data()
````

## Task 1.2

Write a function `get_books_by_genre(books, genre)` to:
- take the list of books, as returned by the function from **Task 1.1**, and a
genre STRING as input arguments
- return a list of book titles that belong to the specified genre.
<div style="text-align: right">[5]</div>

Test your program with *“Fiction”* and show the returned list.
<div style="text-align: right">[1]</div>

````python
# Task 1.2

Book = read_books_data()

print (Book)

def get_books_by_genre (Books, genre):
    list_books = []
    for i in Books:
        if i[1] == genre:
            list_books.append(i[0])
    return list_books

get_books_by_genre(Book, 'Fiction')
````

## Task 1.3

Write a function `get_expensive_books(books, price_threshold)` to:
- take the list of books, as returned by the function from Task 1.1, and a
price threshold DECIMAL as input arguments
- return a list of book titles that have a price greater than or equals to the
specified price threshold.
<div style="text-align: right">[1]</div>

Test your program with *15.99* and show the returned list.
<div style="text-align: right">[1]</div>

````python
def get_expensive_books (books, price_threshold):
    list_below = []
    for i in books:
        if float(i[2]) >= price_threshold:
            list_below.append(i[0])
    return list_below

books = read_books_data()
get_expensive_books (books, 15.99) 
````
