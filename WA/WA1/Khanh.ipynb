````python
# Task 2
# [1] two loops to iterate through grid
# [1] correct logic for comparison, grid[i][j] == letter
# [0] all correct test cases
#       - Your function doesn't return the list object as required by the question.
grid = [
    ['T','I','E','L','C','E'],
    ['T','E','U','C','A','C'],
    ['L','D','R','T','R','I'],
    ['O','E','H','O','D','D'],
    ['C','L','U','E','C','L'],
    ['H','C','T','A','M','S'],
]

# print(len(grid))
def first_letter_search(letter, grid):
    position = []
    x=len(grid)
    for i in range (x):
        for j in range (x):
            if grid[i][j] == letter:
                find = [i,j]
                position.append(find)

    print(position) #correct must be return here, not print


first_letter_search('C', grid)
            
````
