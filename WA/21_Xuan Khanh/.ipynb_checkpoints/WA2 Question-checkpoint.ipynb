# Question 1
Write pseudocode for a program that:
- ask the user for three inputs (all integers)
- outputs the second largest integer

You are **not** allowed to call on any sort function in your program.

You can assume that all the integers are different.

<div style="text-align: right">[3]</div>

````python
FUNCTION SecondLargest (no1: INTEGER, no2: INTEGER, no3: INTEGER) -> INTEGER:
    INTEGER largest <- no1
    INTEGER second_largest <- no2
    if no2 > largest
        THEN
        second_largest <- largest
        largest <- no2
    ENDIF
    
    if no3 > second_largest
        if no3 > largest
        THEN
            second_largest <- largest
            largest <- no3
        ENDIF
        
        second_largest <- no3
        
    ENDIF
    
    return second_largest
        


        
````

One common technique is to replace the missing values with the median value.

The median of a sorted list of values is the central value that divides the list into two equal halves. If the number of elements $n$ in the list is odd, the median is the value at the index $\dfrac{n-1}{2}$. If $n$ is even, then the median is the average of the values at the indices $\dfrac{n-2}{2}$ and $\dfrac{n}{2}$.
