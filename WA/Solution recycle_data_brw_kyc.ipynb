# Task 3

Write a function `task_3()` that:
-  has two parameters:
    - a float `f` that represents a recycling rate
    - a list of list `[[<waste_type>, <waste_disposed_of_tonne>, <total_waste_recycled_tonne>, <total_waste_generated_tonne>, <recycling_rate>, <year>]]` where each sublist is a copy of a row of record from the file `2003_2017_waste.csv`.
- returns the most common waste type (as a string) amongst waste types with recyling rates lesser than `f`. 
- If there is a tie for the most common waste type, return any one of the waste type that is the most common.
- The lowest recycling rate is 0.0 and the highest is 1.0. 
- The function should include the ability to handle <b>erroneous</b> input from the user.

<div style="text-align: right">[5]</div>

````python
# Task 3
# [1] - try except or isinstance(f, float)
# [1] - range check for value of f
# [1] - find all waste types with recycling rates lesser than f
# [1] - create the dictionary or equivalent method to find waste type of highest frequency
# [1] - correctly find the highest frequency waste type for recycling rate below f

waste_list = task_1('2003_2017_waste.csv')

def task_3(f, waste_list):
    if not (isinstance(f,float) or isinstance(f,int)):
        return 'input has to be a float'
    else:
        if (f>1.0) or (f<0.0):
            return 'input has to be numeric between 0.0 and 1.1'
        else:
            filtered = [w for w in waste_list if float(w[4])<f]
            waste_type_filtered = list(set([w[0] for w in filtered]))
            waste_type_freq = {}
            for w in waste_type_filtered:
                waste_type_freq[w] = 0
            for w in filtered:
                waste_type_freq[w[0]] += 1
            waste_type_freq_list = [[k,v] for k,v in waste_type_freq.items()]
            waste_type_freq_list = sorted(waste_type_freq_list, key = lambda x:x[1], reverse = True)
    print(waste_type_freq_list)
    return waste_type_freq_list[0]


waste_list = task_1('2003_2017_waste.csv')
print(task_3(0.05,waste_list))
````

````python
def task_3(f: float)->'str':
    try:
        f = float(f) # ValueError and TypeError handling [1]
        if f > 1: # range check
            return 'Invalid Recycling Rate'
        elif f == 0:
            return None
        else:
            data = task_1('2003_2017_waste.csv')
            waste_lt_rr = [d[0] for d in data if float(d[4]) < f]
            waste_lt_rr_set = set(waste_lt_rr)
            waste_lt_rr_count_dict = {w:waste_lt_rr.count(w) for w in waste_lt_rr_set}
            max_waste_count = max(list(waste_lt_rr_count_dict.values()))
            # print(waste_lt_rr_count_dict) 
            # print(max_waste_count)
            for waste, count in waste_lt_rr_count_dict.items():
                if count == max_waste_count:
                    return waste
    except ValueError:
        print('ValueError! input cannot be converted to float')
    except TypeError:
        print('TypeError, expecting integer, float or string')

task_3(0.5)
````

# Task 5
The 2008 data is missing from the list generated by `task_1()`. In such situation of missing data, we can replace missing values in a dataset with substituted, estimated values. This process is called `data imputation`.

One common technique is to replace the missing values with the median value.

The median of a sorted list of values is the central value that divides the list into two equal halves. If the number of elements $n$ in the list is odd, the median is the value at the index $\dfrac{n-1}{2}$. If $n$ is even, then the median is the average of the values at the indices $\dfrac{n-2}{2}$ and $\dfrac{n}{2}$.

The missing recycling rate for each waste type in the year 2008 can be estimated with the median recycling rate for the respective waste type of the available years from the dataset.

Write a function `task_5()` that takes in the list with the missing data from 2008 and returns a list of tuple containing the estimated recycling rate for the year 2008 of each waste type. The list should be of the form 

```text
[(<waste_type_1>, <recycling_rate_1>), (<waste_type_2>, <recycling_rate_2>),...].  
```

You should not be importing modules to find the median recycling rate. 


<div style="text-align: right">[5]</div>

````python
# [1] - create a dictionary as waste type as key...
# [1]      - ... and correctly store the values for recycling rate for corresponding keys, or equivalent
# [1] - correct conditions implemented when calculating for median....
# [1]   - ... with correct returns for different cases
# [1] - return correct list of tuples of (<waste_type>, <median>)
#       - the indices to get the middle of the array might not be an integer in your implementation, use floor division

waste_list = task_1('2003_2017_waste.csv')

def task_5(waste_list):
    def medi(lst):
        size = len(lst)
        if len(lst) % 2 == 0:
            return (1/2)*(lst[size//2] + lst[(size-2)//2])
        else:
            return lst[(size-1)//2]
    waste_type = list(set([w[0] for w in waste_list]))
    waste_type_rec_rate = {}
    for w in waste_type:
        waste_type_rec_rate[w] = []
    for w in waste_list:
        waste_type_rec_rate[w[0]].append(float(w[4]))
    approx_rate = []
    for k,v in waste_type_rec_rate.items():
        approx_rate.append((k,medi(v)))
    return approx_rate

task_5(waste_list)
````

````python
def find_median(r_list):
    if len(r_list) % 2 == 0:
        mid1 = (len(r_list) // 2) -1
        mid2 = len(r_list) // 2
        return (float(r_list[mid1]) + float(r_list[mid2]))/2
    else:
        return r_list[len(r_list)//2]


def task_5(r_data: list):
    waste_rr = {}
    for record in r_data:
        waste_type = record[0]
        recycling_rate = record[4]
        if waste_type not in waste_rr:
            waste_rr[waste_type] = [recycling_rate]
        else:
            waste_rr[waste_type].append(recycling_rate)
    print(waste_rr)
    missing_rates = []
    for waste, rates in waste_rr.items():
        median = find_median(rates)
        missing_rates.append((waste, median))
    return missing_rates

task_5(task_1('2003_2017_waste.csv'))
````
