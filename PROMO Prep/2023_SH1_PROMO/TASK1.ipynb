```python
def task1_1(filename):
    with open(filename) as f:
        file = f.read()
        list = [int(i) for i in file.split()]
    return list

result = task1_1("TEN.txt")
print(result)
print(len(result))
```

```python
def task1_2(list): #bubble sort
    n = len(list)
    swapped = False
    for i in range (n):
        for j in range (0,n-1):
            if list[j] > list[j+1]:
                list[j], list[j+1] = list[j+1], list[j]
                swapped = True
        
        if swapped == False:
            break
    return list

arr = [8, 5, 2, 1, 3, 7, 9]
task1_2 (arr)

print(task1_2(task1_1("TEN.txt")))
```

```python
import random
def task1_3(list): #quicksort
    if len(list) <= 1:
        return list
    left = []
    middle = []
    right = []

    index = random.randint(0,len(list)-1)
    for element in list:
        if element < list[index]:
            left.append(element)
        elif element > list[index]:
            right.append(element)
        else:
            middle.append(element)

    return task1_3(left) + task1_3(middle) + task1_3(right)

arr = [8, 5, 2, 1, 3, 7, 9]
print(task1_3(task1_1('TEN.txt')))
```
