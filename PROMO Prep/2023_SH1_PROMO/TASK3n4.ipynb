```python
import math
import random

class Player:
    def __init__(self, name, strength, agility, stamina):
        self.name = name
        self.st= float(strength)
        self.agi = float(agility)
        self.stam = float(stamina)
    
    def set_name(self,new_name):
        self.name = new_name
    def set_st(self,new_st):
        self.st = new_st
    def set_agi(self,new_agi):
        self.agi = new_agi
    def set_stam(self,new_stam):
        self.stam = stam
    
    def get_name(self):
        return self.name
    def get_st(self):
        return self.st
    def get_agi(self):
        return self.agi
    def get_stam(self):
        return self.stam
    
    def play(self, Opponent):
        w1 = 0.25*self.st + 0.35*self.agi + 0.4*self.stam
        w2 = 0.25*Opponent.st + 0.35*Opponent.agi + 0.4*Opponent.stam
        p = (2.718**w1) / (2.718**w1 + 2.718**w2)
        chance = random.uniform(0,1)
        if chance < p:
            return True
        else:
            return False
    
    def get_chance(self, Opponent): 
        w1 = 0.25*self.st + 0.35*self.agi + 0.4*self.stam
        w2 = 0.25*Opponent.st + 0.35*Opponent.agi + 0.4*Opponent.stam
        p = (2.718**w1) / (2.718**w1 + 2.718**w2)
        return round(p, 3)

p1 = Player("p1",1, 2, 3)
p2 = Player("p2",2,3,4)
print(p1.play(p2))
print(p1.get_chance(p2))
```
