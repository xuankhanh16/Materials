```python
CREATE_TABLE = '''

CREATE TABLE IF NOT EXISTS Game(
    GameID Integer Unique,
    Title TEXT,
    Genre TEXT,
    TimeSinceRelease float,
    Primary Key(GameID)
);

CREATE TABLE IF NOT EXISTS Renter(
    RenterID Integer Unique,
    Name TEXT,
    ContactNo TEXT,
    ContactEmail TEXT,
    Primary Key(RenterID)
);

CREATE TABLE IF NOT EXISTS Rental(
    RentalID Integer,
    RenterID integer,
    GameID integer,
    DateLoaned TEXT,
    Returned bool,
    Primary Key(RentalID),
    Foreign Key(RenterID) REFERENCES Renter(RenterID),
    Foreign Key (GameID) REFERENCES Game(GameID)
)

'''

```

```python
import sqlite3

conn = sqlite3.connect("library.db")

cursor = conn.cursor()

cursor.executescript(CREATE_TABLE)
conn.commit()

```

The text files GAME.csv, RENTER.csv and RENTAL.csv store the comma-separated values
for each of the tables in the database.
Write a Python program to read in the data from each file and then store each item of
data in the correct place in the database

```python
#Task 5.2
import csv
import sqlite3


conn = sqlite3.connect("library.db")
cursor = conn.cursor()


with open("Game.csv") as f:
    file = csv.reader(f)
    list_game = [i for i in file]
    list_game = list_game[1:]
    for game in list_game:
        id = int(game[0])
        name = game[1]
        type = game[2]
        time = game[3]
        print(id, name, type, time)
        cursor.execute("INSERT INTO GAME (GameID, Title, Genre, TimeSinceRelease) VALUES (?,?,?,?)", (id, name, type, time))
        conn.commit()
    print(list_game)

with open ("RENTER.csv") as f:
    file = csv.reader(f)
    list_renter = [i for i in file]
    list_renter = list_renter[1:]
    print(list_renter)
    for renter in list_renter:
        id = int(renter[0])
        name = renter [1]
        contact = renter [2]
        email = renter[3]
        cursor.execute("INSERT INTO RENTER(RenterID, Name, ContactNo, ContactEmail) VALUES (?,?,?,?)", (id, name, contact, email))
        conn.commit()

with open ("RENTAL.csv") as f:
    file = csv.reader(f)
    list_rental = [i for i in file]
    list_rental = list_rental[1:]
    print(list_rental)
    for rental in list_rental:
        rental_id = int(rental[0])
        renter_id = int(rental [1])
        GameID = int(rental[2])
        DateLoaned = rental[3]
        returned = rental[4]
        cursor.execute("INSERT INTO RENTAL(RentalID, RenterID, GameID, DateLoaned, Returned) VALUES (?,?,?,?,?)", (rental_id, renter_id, GameID, DateLoaned, returned))
        conn.commit()
```

```python
from flask import Flask, render_template
import sqlite3

conn = sqlite3.connect('library.db')
cursor = conn.cursor()

popularity_index = '''
    SELECT
        G.GameID,
        G.Title,
        (COUNT(R.RentalID) * 1.0) / G.TimeSinceRelease AS MPOP 
    FROM
        Game AS G
    LEFT JOIN
        Rental AS R ON G.GameID = R.GameID
    GROUP BY
        G.GameID, G.Title, G.TimeSinceRelease
    ORDER BY
        MPOP DESC;
    '''

cursor.execute(popularity_index)
list_mop = [i for i in cursor]

app = Flask(__name__)

@app.route("/")
def display():
    return render_template("index.html", list = list_mop)

app.run()
```
