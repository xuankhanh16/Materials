The result of an international bowling tournament is stored in a file named `HEX_SCORE.TXT`. Due to a glitch in the scoring system, the values of the score are all in hexadecimal numbers. The file contains the player ID, the country code for the country the player is representing and the scores for the six games played in the tournament.

## Task 3.1

Write Python code to convert a string containing hexadecimal number to denary numbers. You are not allowed to use any built-in Python function.
<div style='text-align: right; font-family: Courier New, monospace;'>[5]</div>

```python
# Task 3.1

def hex_to_denary (string):
    sum = 0
    count = 0
    for i in string[::-1]:
        if i.lower() == "a":
            value = 10
        elif i.lower() == "b":
            value = 11
        elif i.lower() == "c":
            value = 12
        elif i.lower() == "d":
            value = 13
        elif i.lower() == "e":
            value = 14
        elif i.lower() == "f":
            value = 15
        else:
            value = int(i)

        sum += value * (16**count)
        count += 1
    return sum

string = input()
hex_to_denary (string)
```

```python
# Task 3.2

import csv

def bubble_sort(arr):
    n = len(arr)
    for i in range (n-1):
        swapped = False
         
        for j in range (n-i-1):
            if arr[j][1] < arr[j+1][1]: #to inverse sort
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
            
        if not swapped:
            break
        
    return arr


with open("HEX_SCORES.txt") as f:
    file = csv.reader(f)
    selected_result = []
    result = [i for i in file]
    for person in result:
        sum_person = 0
        for i in range (2, 8):
            sum_person += hex_to_denary (person[i])
        selected_result.append([person[0], sum_person])
    
    #sorted_selected_result = sorted(selected_result, key = lambda x: x[1], reverse = True)
    sorted_selected_result = bubble_sort(selected_result)

print(f"Rank Id Score \n ")
for i in range (1, len(sorted_selected_result)+1):
    print(f"{i} {sorted_selected_result[i-1][0]} {sorted_selected_result[i-1][1]}")



x = bubble_sort(selected_result)
#print(sorted_selected_result)
print(x)
```
