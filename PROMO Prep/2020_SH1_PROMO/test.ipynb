```python
#Task 1.1

import csv

def read_data (filename):
    filepath = "resources/" + filename
    with open(filepath) as f:
        reader = csv.reader(f, delimiter = ";")
        i = [x for x in reader]
    
    return i

read_data("PARTICIPANTS.txt")
```

```python
#Task 1.2

participants = read_data("PARTICIPANTS.txt")
participants

def gender_count(participants):
    male_cnt = 0
    female_cnt = 0
    for i in participants:
        if i[2] == "F":
            female_cnt += 1
        else: 
            male_cnt +=1
    return (male_cnt, female_cnt)

male, female = gender_count(participants)
print(male, female)
```

```python
#Task 1.3


def role_statistics (participants):
    role = {}
    for i in participants:
        my_role = i[1]
        if my_role in role:
            role[my_role] += 1
        else:
            role[my_role] = 1
    
    print (f"Role" + "   " + "Number")
    for the_role, number in role.items():
        print(f"{the_role}  {number}")

role_statistics (participants) 
```

```python
# Task 2.2

def Fibonacci (n):
    if (n<=0):
        return 0
    elif(n<=2):
        return 1

    number = 1
    prev_number = 1
    for i in range (2,n):
        number_dummy = number
        number = prev_number + number
        prev_number = number_dummy

    return number

x = int (input())

for i in range (1,x+1):
    print(Fibonacci(i))
        
        

```

```python
#Task 3.1

class Visitor:

    def __init__(self, name, contactNumber):
        self._name = name
        self._contactNumber = contactNumber
    
    def getName (self):
        return self._name
    
    def getContact (self):
        return self._contactNumber

    def __str__(self):
        return (f"{self.getName()} : {self.getContact()}")


class Patient:
    def __init__ (self, name, nric):
        self._name = name
        self._nric = nric
    
    def getName (self):
        return self._name
    
    def getNRIC (self):
        return self._nric
    
    def __str__(self):
        return (self.getName())
    

```

```python
#Task 3.3

class Queue:
    def __init__ (self):
        self._buffer = []
    
    def enqueue (self, Visitor):
        if len(self._buffer) < 10:
            self._buffer.append(Visitor)
            return True
        else:
            return False

    def dequeue (self):
        if not self._buffer: #if self._buffer is empty
            return None
        else:
            self._buffer.pop(0)
    
    def __str__ (self):
        return f"[{','.join(str(visitor) for visitor in self._buffer)}]"



```

```python
# Task 3.4
queue_obj = Queue()
queue_obj.enqueue(visitor1)
queue_obj.enqueue(visitor2)
print(queue_obj)

queue_obj.dequeue()
print(queue_obj)

visitor3 = Visitor("Alvinia", "955-561-84")
queue_obj.enqueue(visitor3)
print(queue_obj)
```
