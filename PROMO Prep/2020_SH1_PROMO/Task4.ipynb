```python
query = [
    '''
CREATE TABLE School (
    SchoolCode Integer Unique,
    Name Text,
    Address Text,
    Primary Key(SchoolCode)
)
    ''',
    '''
CREATE Table Staff(
    SchoolCode Integer ,
    Name TEXT ,
    Department TEXT,
    Contact TEXT,
    Foreign Key (SchoolCode) references School(SchoolCode)
)
    '''
]
```

```python
#TASK 4.1
import sqlite3

conn = sqlite3.connect("schools.db")
cursor = conn.cursor()

for i in query:
    cursor.execute(i)
    
```

```python
#Task 4.2
import csv
import sqlite3

conn = sqlite3.connect("schools.db")
cursor = conn.cursor()

def task4_2():
    #import school db
    with open("resources/School.txt") as f:
        result = list(i for i in csv.reader(f))[1:]
        print(result)
    import_sch_query = '''
    INSERT INTO School(SchoolCode, Name, Address) VALUES (?,?,?)
        '''
    for school in result:
        cursor.execute(import_sch_query, school)


    #import staff db
    with open("resources/Staff.txt") as f:
        staff_result = list(x for x in csv.reader(f))[1:]
        print(staff_result)

    import_staff_query = '''
INSERT INTO Staff (SchoolCode, Name, Department, Contact) VALUES (?,?,?,?)
    '''

    for staff in staff_result:
        cursor.execute(import_staff_query, staff)
    conn.commit()

task4_2()
```

```python
#Task 4.3

import sqlite3

conn = sqlite3.connect("schools.db")
cursor = conn.cursor()

display_query = ''' 
SELECT
    sch.Name,
    st.Name,
    st.Department,
    st.contact
FROM 
    school sch
INNER JOIN
    staff st ON sch.SchoolCode = st.SchoolCode
'''

cursor.execute(display_query)
data = cursor.fetchall()

```

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route("/")
def home():
    return render_template("home.html", list = data)

app.run()
```
